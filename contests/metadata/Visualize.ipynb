```python
import yaml
```

```python
with open("contexts.yaml", "r") as f:
    contexts = yaml.safe_load(f)

with open("anomalies.yaml", "r") as f:
    anoms = yaml.safe_load(f)

assert contexts.keys() == anoms.keys()

for k, context in contexts.items():
    anom = anoms[k]
    same = set(context).intersection(anom)
    if same:
        print(k, same)
```

```python
{k: len(v) for k, v in anoms.items() if len(v) <= 1 or len(v) >= 5}
```

```python
import pandas as pd
pd.Series([len(x) for x in contexts.values()]).plot.hist()
```
